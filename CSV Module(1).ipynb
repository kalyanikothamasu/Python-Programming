```python
import csv

student_data = [
    {'student_id': 1, 'name': 'Alice', 'age': 20, 'major': 'Computer Science'},
    {'student_id': 2, 'name': 'Bob', 'age': 22, 'major': 'Physics'},
    {'student_id': 3, 'name': 'Charlie', 'age': 21, 'major': 'Mathematics'},
    {'student_id': 4, 'name': 'David', 'age': 19, 'major': 'Engineering'},
    {'student_id': 5, 'name': 'Eve', 'age': 23, 'major': 'Biology'}
]

# You can display the data to see the generated list of dictionaries
# display(student_data)

with open('student.csv','w') as student_file:
    csvwriter = csv.writer(student_file)
    # Assuming you want to write the data as rows in the CSV
    # First write the header
    if student_data:
        csvwriter.writerow(student_data[0].keys())
        # Then write the data rows
        for row in student_data:
            csvwriter.writerow(row.values())

print("Data written to student.csv")
```

```python
import csv

student_data_list = [
    [1, 'Alice', 20, 'Computer Science'],
    [2, 'Bob', 22, 'Physics'],
    [3, 'Charlie', 21, 'Mathematics'],
    [4, 'David', 19, 'Engineering'],
    [5, 'Eve', 23, 'Biology']
]

# Write data to a CSV file
with open('student_list.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    # Write header (optional, based on your list of lists structure)
    writer.writerow(['student_id', 'name', 'age', 'major']) # Example header
    writer.writerows(student_data_list)

print("Data written to student_list.csv")

# Read data from the CSV file using csv reader
print("\nReading data from student_list.csv:")
with open('student_list.csv', 'r') as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```
