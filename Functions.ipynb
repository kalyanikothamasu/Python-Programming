```python
import zipfile
import os

def compress_file(file_path, zip_path):
  """Compresses a single file into a zip archive.

  Args:
    file_path: The path to the file to compress.
    zip_path: The path for the output zip archive.
  """
  try:
    with zipfile.ZipFile(zip_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
      zipf.write(file_path, os.path.basename(file_path))
    print(f"Successfully compressed {file_path} to {zip_path}")
  except FileNotFoundError:
    print(f"Error: The file {file_path} was not found.")
  except Exception as e:
    print(f"An error occurred during compression: {e}")

# Example usage:
# Create a dummy file to compress
with open('my_test_file.txt', 'w') as f:
  f.write("This is a test file for compression.")

compress_file('my_test_file.txt', 'my_test_archive.zip')
```

```python
# Sample dictionary
my_dict = {'song': 'Malli puyeve', 'Movie': 'pushparaj', 'year': 2023}

# Get the value of an existing key
song_title = my_dict.get('song', 'Unknown Song')
print(f"Song title: {song_title}")

# Get the value of a non-existing key with a default value
director = my_dict.get('director', 'Unknown Director')
print(f"Director: {director}")
```

```python
# Sample list of dictionaries (some might be missing the 'city' key)
locations = [
    {'name': 'Alice', 'city': 'New York'},
    {'name': 'Bob', 'city': 'London'},
    {'name': 'Charlie'}, # Missing 'city' key
    {'name': 'David', 'city': 'Paris'}
]

# Use list comprehension with get() to get the city for each person, with a default
cities = [item.get('city', 'Unknown City') for item in locations]

# Print the list of cities
print(cities)
```
