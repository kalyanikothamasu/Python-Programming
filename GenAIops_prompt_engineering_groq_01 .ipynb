````python
import pandas as pd
import random
csv_file_path='/content/sample_data/california_housing_test.csv'

def generate_prompt(csv_file_path, num_samples=60):
    """Generates a prompt with a static section and random examples from a CSV."""

    static_prompt = """
#role:
You are a Machine learning expert who can predict home prices based on few examples. Speculative predictions are okay for now

#instruction:
Task: Predict the price of a home based on the provided features.

#additinoal context:
Input: A list of features describing a home
Output: The predicted price of the home.

#few Examples below:
"""

    try:
        df = pd.read_csv(csv_file_path)
    except FileNotFoundError:
        return f"Error: CSV file not found at {csv_file_path}"
    except pd.errors.EmptyDataError:
        return f"Error: CSV file is empty: {csv_file_path}"
    except Exception as e:
        return f"Error reading CSV file: {e}"

    if num_samples > len(df):
        return f"Error: Requested {num_samples} samples exceeds available {len(df)} rows."

    sampled_df = df.sample(n=num_samples, replace=False)

    examples = []
    for i, row in sampled_df.iterrows():  # Use i for example number
        input_str = "Input:\n"
        for col in df.columns:
            if col != "median_house_value":
                input_str += f"{col}: {row[col]}\n"

        output_str = f"Output:\nmedian_house_value: {row['median_house_value']}\n\n"
        examples.append(f"Example {i + 1}:\n{input_str}{output_str}")

    return static_prompt + "".join(examples)


# Example usage:
csv_file_path = "/content/sample_data/california_housing_train.csv"  # Replace with your path
prompt = generate_prompt(csv_file_path)

if "Error" not in prompt:
    print(prompt)
    try:
        with open("generated_prompt.txt", "w") as f:
            f.write(prompt)
        print("Prompt saved to generated_prompt.txt")
    except Exception as e:
        print(f"Error saving prompt: {e}")
else:
    print(prompt)
````
