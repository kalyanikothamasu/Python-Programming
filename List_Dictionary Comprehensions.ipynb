```python
# Demonstrating the use of conditional expressions (ternary operator) in lambda functions

# Example 1: Assigning a value based on a condition
# If the number is even, return 'Even', otherwise return 'Odd'
parity_checker = lambda x: 'Even' if x % 2 == 0 else 'Odd'

print("Parity of 7:", parity_checker(7))
print("Parity of 10:", parity_checker(10))

# Example 2: Using a conditional expression within map()
# Square the number if it's even, otherwise keep it as is
numbers = [1, 2, 3, 4, 5]
processed_numbers = list(map(lambda x: x**2 if x % 2 == 0 else x, numbers))
print("Processed numbers:", processed_numbers)
```

```python
# From a list of strings data = ['name=Alice', 'age=30', 'city=Boston'], create a dictionary from the key-value pairs.

data = ['name=Alice', 'age=30', 'city=Boston']

# Using dictionary comprehension
data_dict_comp = {item.split('=')[0]: item.split('=')[1] for item in data}
print("Dictionary created from strings (comprehension):", data_dict_comp)

# Another method using dict() and a generator expression
data_dict_dict_gen = dict(item.split('=') for item in data)
print("Dictionary created from strings (dict + generator):", data_dict_dict_gen)

# Note: These methods assume each string contains exactly one '=' and there are no issues with values containing '='.
# For more complex cases, more robust parsing might be needed.
```

```python
sentence = "Hello World"

# Method 1: Create a list of tuples (lowercase, uppercase) for each letter
# This will only include letters and ignore spaces or other characters.
case_pairs = [(i.islower(), i.isupper()) for i in sentence if i.isalpha()]
print(case_pairs)

# Method 2: Create a nested list containing lowercase and uppercase lists
# This might not be exactly what "at a time" means, but it uses one comprehension
# to generate a structure containing both.
nested_cases = [[i.lower() for i in sentence if i.isalpha()], [i.upper() for i in sentence if i.isalpha()]]
print(nested_cases)

# Method 3: Using a conditional expression to output either lower or upper based on some criteria (less likely for "both at a time" but demonstrates conditional output)
# For example, if you wanted lowercase for letters and the original character otherwise:
mixed_case = [i.lower() if i.isalpha() else i for i in sentence]
print(mixed_case)

# Keep in mind that a single list comprehension inherently builds a single list.
# Generating two separate lists simultaneously within one list comprehension is not a direct capability.
```
