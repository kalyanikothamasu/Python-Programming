```python
import pandas as pd
import numpy as np

# Example 1: Handling missing values with mean()
df1 = pd.DataFrame({'col1': [1, 2, np.nan, 4, 5]})
print("Example 1: Mean with NaN")
print(df1['col1'].mean())

# Example 2: Handling missing values with sum()
print("\nExample 2: Sum with NaN")
print(df1['col1'].sum())

# Example 3: Filling missing values with fillna()
print("\nExample 3: Filling NaN with 0")
print(df1['col1'].fillna(0))

# Example 4: Dropping missing values with dropna()
print("\nExample 4: Dropping NaN")
print(df1['col1'].dropna())

# Example 5: Using value_counts() with NaN
df5 = pd.DataFrame({'col1': ['A', 'B', 'A', np.nan, 'B', 'A']})
print("\nExample 5: value_counts() with NaN")
print(df5['col1'].value_counts(dropna=False))

# Example 6: Concatenating DataFrames with different columns
df6_1 = pd.DataFrame({'colA': [1, 2], 'colB': [3, 4]})
df6_2 = pd.DataFrame({'colC': [5, 6], 'colD': [7, 8]})
print("\nExample 6: Concatenating DataFrames with different columns")
print(pd.concat([df6_1, df6_2], axis=1))

# Example 7: Concatenating DataFrames with overlapping columns
df7_1 = pd.DataFrame({'colA': [1, 2], 'colB': [3, 4]})
df7_2 = pd.DataFrame({'colA': [5, 6], 'colC': [7, 8]})
print("\nExample 7: Concatenating DataFrames with overlapping columns")
print(pd.concat([df7_1, df7_2], join='inner'))

# Example 8: Merging DataFrames with different column names for the merge key
df8_1 = pd.DataFrame({'employee_id': [1, 2], 'Name': ['A', 'B']})
df8_2 = pd.DataFrame({'id': [1, 2], 'Salary': [100, 200]})
print("\nExample 8: Merging with different key names")
print(pd.merge(df8_1, df8_2, left_on='employee_id', right_on='id'))

# Example 9: Merging DataFrames with duplicate key values
df9_1 = pd.DataFrame({'colA': [1, 2, 2], 'colB': ['A', 'B', 'C']})
df9_2 = pd.DataFrame({'colA': [1, 2], 'colC': ['X', 'Y']})
print("\nExample 9: Merging with duplicate keys")
print(pd.merge(df9_1, df9_2, on='colA'))

# Example 10: Using groupby() on an empty DataFrame
df10 = pd.DataFrame(columns=['col1', 'col2'])
print("\nExample 10: Groupby on empty DataFrame")
print(df10.groupby('col1').sum())

# Example 11: Using apply() with a function that returns different types
df11 = pd.DataFrame({'col1': [1, 2, 3]})
print("\nExample 11: apply() with mixed return types")
print(df11['col1'].apply(lambda x: str(x) if x % 2 == 0 else x))

# Example 12: Using pivot_table() with duplicate entries
df12 = pd.DataFrame({'colA': ['A', 'A', 'B'], 'colB': ['X', 'X', 'Y'], 'colC': [1, 2, 3]})
print("\nExample 12: pivot_table() with duplicate entries")
print(df12.pivot_table(index='colA', columns='colB', values='colC', aggfunc='sum'))

# Example 13: Using sort_values() with missing values
df13 = pd.DataFrame({'col1': [1, np.nan, 3, 2], 'col2': ['A', 'B', 'C', 'D']})
print("\nExample 13: sort_values() with NaN")
print(df13.sort_values(by='col1'))

# Example 14: Using sort_index() with a non-standard index
df14 = pd.DataFrame({'col1': [1, 2]}, index=['C', 'A'])
print("\nExample 14: sort_index() with non-standard index")
print(df14.sort_index())

# Example 15: Using rename() with a non-existent column
df15 = pd.DataFrame({'col1': [1, 2]})
print("\nExample 15: rename() with non-existent column")
print(df15.rename(columns={'non_existent_col': 'new_name'}))

# Example 16: Using drop() with a non-existent column
df16 = pd.DataFrame({'col1': [1, 2]})
# This will raise an error if errors='raise' (default)
# print("\nExample 16: drop() with non-existent column")
# print(df16.drop(columns=['non_existent_col']))
print("\nExample 16: drop() with non-existent column (errors='ignore')")
print(df16.drop(columns=['non_existent_col'], errors='ignore'))


# Example 17: Using unique() on a column with NaN
df17 = pd.DataFrame({'col1': ['A', 'B', 'A', np.nan, 'B', 'A']})
print("\nExample 17: unique() with NaN")
print(df17['col1'].unique())

# Example 18: Using nunique() on a column with NaN
print("\nExample 18: nunique() with NaN")
print(df17['col1'].nunique(dropna=False))
print(df17['col1'].nunique(dropna=True))


# Example 19: Using replace() with multiple values
df19 = pd.DataFrame({'col1': ['A', 'B', 'A', 'C', 'B']})
print("\nExample 19: replace() with multiple values")
print(df19['col1'].replace(['A', 'B'], ['X', 'Y']))

# Example 20: Using replace() with a dictionary and regex
df20 = pd.DataFrame({'col1': ['A1', 'B2', 'C3']})
print("\nExample 20: replace() with dictionary and regex")
print(df20['col1'].replace({'[A-Z]': 'Letter'}, regex=True))
```
