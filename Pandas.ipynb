## Pandas

## What is Pandas?

1. pandas is a fast, powerful,open-source **python Library** used for **Data Manipulations and Analysis**
2. It Provides easy-to-use Data Structures like
    - Series (1-Dimensional labeled array)
    - DataFrame( 2 -Dimensional labeled Data Table, like Excel,CSV files)

## Why to use Pandas?
1. Read and write data from different file formats (CSV,Excel,JSON,etc)
2. Clean and prepare messy datasets
3. Perform Filtering, Grouping, merging and reshaping operations with simple conditions
4. Handle Missing Values ,duplicates, and large datasets effectively

## Advantages of pandas
1. User Friendly
2. Handles large data
3. Data Cleaning
4. Integration - works well with numpy, matplotlib,scikit-learn, tensorflow,etc
5. Powerful Data Operations - Filtering, Grouping, Joining, Merging, Pivoting, and Time Series analysis
6. Speed & Performance - Optimized for performance using C
7. Visualization Support - Can quickly plot graphs using .plot() with matplotlib

### Value_counts( )
- **value_counts()** is Method in Pandas used on **Series**( a single column in a DataFrame)
- It returns the **count of unique values** in that column, sorted by default in descending order

```python
df = pd.DataFrame({'Fruits':['Apple','Banana','Apple','Mango','Apple','Apple','Mango','Banana','Apple','Mango'],
                  'Numbers':[1,3,4,1,2,3,4,5,1,3],
                  'Grades':['A','A','B','C','A','B','A','C','A','C']})
```

```python
df
```

```python
df['Fruits'].value_counts()
```

#### Get percentages instead of counts

```python
df['Fruits'].value_counts(normalize=True)
```

#### include NaN values

```python
df_with_nan = pd.Series(['A','B','A',None,'B',None,'A','B','C','B'])
```

```python
df_with_nan
```

```python
df_with_nan.value_counts(dropna=False)
```

```python
df_with_nan.value_counts()
```

#### Sorting in Acending Order

```python
df['Grades'].value_counts(ascending=True)
```

#### Binning continuous Values

```python
df['Numbers']
```

```python
df['Numbers'].value_counts(bins=3)
```

### drop( )
- The **drop()** Method is used to remove **rows or Columns** from Data Frame
- It Does not change the DataFrame by default unless you use **inplace = True**

```python
df1.drop('Unnamed: 15',axis=1) # axis = 1 ---> columns
```

```python
df1.drop('Unnamed: 15',axis=1,inplace=True)
```

```python
df1
```

```python
df1.info()
```

```python
df1.isna().sum() # sum of null values in each column
```
