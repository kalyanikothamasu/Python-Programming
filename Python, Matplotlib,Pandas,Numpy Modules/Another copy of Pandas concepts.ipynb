```python
import pandas as pd
import numpy as np
import datetime as dt

# Generate 1000 random dates within a two-year period
start_date = pd.to_datetime('2023-01-01')
end_date = pd.to_datetime('2024-12-31')
date_range = pd.date_range(start=start_date, end=end_date, periods=1000)

# Generate random spending amounts
spending_amounts = np.random.uniform(10, 500, 1000).round(2)

# Define spending categories and locations
categories = ['Groceries', 'Electronics', 'Clothing', 'Home', 'Entertainment', 'Other']
locations = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix', 'Philadelphia', 'San Antonio', 'San Diego', 'Dallas', 'San Jose']

# Generate random categories and locations for each date
random_categories = np.random.choice(categories, 1000)
random_locations = np.random.choice(locations, 1000)

# Create the DataFrame
spending_df = pd.DataFrame({
    'Date': date_range,
    'Category': random_categories,
    'Spend': spending_amounts,
    'Location': random_locations
})

# Display the first few rows of the DataFrame
display(spending_df.head())

# Extract the date part from the 'Date' column
spending_df['Date_only'] = spending_df['Date'].dt.date
display(spending_df.head())
spending_df.drop(columns=["Date"])
```

# Task
Calculate the day-to-day and month-to-month percentage change in spending from the `spending_df` DataFrame.

## Aggregate spending by day

### Subtask:
Group the `spending_df` DataFrame by the 'Date_only' column and sum the 'Spend' for each day.


**Reasoning**:
Group the spending_df DataFrame by 'Date_only' and calculate the sum of 'Spend' for each day.



```python
daily_spending = spending_df.groupby('Date_only')['Spend'].sum()
display(daily_spending.head())
```

## Calculate day-to-day percentage change

### Subtask:
Use the `pct_change()` method on the daily spending to find the percentage change from the previous day.


**Reasoning**:
Calculate the day-to-day percentage change of the daily spending using the `.pct_change()` method and display the first few values.



```python
daily_pct_change = daily_spending.pct_change()
display(daily_pct_change.head())
```

## Aggregate spending by month

### Subtask:
Extract the month from the 'Date_only' column, group by month, and sum the 'Spend' for each month.


**Reasoning**:
Extract the month from the 'Date_only' column, group by month, sum the 'Spend', and display the head of the result.



```python
spending_df['Month'] = pd.to_datetime(spending_df['Date_only']).dt.to_period('M')
monthly_spending = spending_df.groupby('Month')['Spend'].sum()
display(monthly_spending.head())
```

## Calculate month-to-month percentage change

### Subtask:
Use the `pct_change()` method on the monthly spending to find the percentage change from the previous month.


**Reasoning**:
Apply the `pct_change()` method to the `monthly_spending` Series to calculate the month-over-month percentage change and display the head of the resulting Series.



```python
monthly_pct_change = monthly_spending.pct_change()
display(monthly_pct_change.head())
```

## Display results

### Subtask:
Display the DataFrames containing the daily and monthly percentage changes.


**Reasoning**:
Print the daily and monthly percentage changes to display the results of the previous calculations.



```python
print("Daily Percentage Change in Spending:")
print(daily_pct_change)

print("\nMonthly Percentage Change in Spending:")
print(monthly_pct_change)
```
