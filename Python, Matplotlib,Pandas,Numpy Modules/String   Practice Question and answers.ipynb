```python
# Find All Occurrences of a Substring: Find all starting indices of a given substring.

main_string = "this is Datateach practice and discussion class and discussion is good"
substring = input("Enter the substring to find: ")

indices = []
start_index = 0

while start_index < len(main_string):
    index = main_string.find(substring, start_index)
    if index == -1:
        break
    indices.append(index)
    start_index = index + 1

if indices:
    print(f"The substring '{substring}' was found at indices: {indices}")
else:
    print(f"The substring '{substring}' was not found in the main string.")
```

```python
# Find Substring Index: Find the starting index of a substring.

main_string = "This is a sample string."
substring = "sample"

# Using find() - returns -1 if not found
find_index = main_string.find(substring)
print(f"Using find(): The substring '{substring}' is found at index {find_index}")

# Using index() - raises a ValueError if not found
try:
    index_index = main_string.index(substring)
    print(f"Using index(): The substring '{substring}' is found at index {index_index}")
except ValueError:
    print(f"Using index(): The substring '{substring}' was not found.")
```
