```python
# prompt: star patterns

n = int(input("Enter the number of rows: "))

# Pattern 1: Increasing numbers in each row
print("\nPattern 1:")
for i in range(1, n + 1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()

# Pattern 2: Decreasing numbers in each row
print("\nPattern 2:")
for i in range(n, 0, -1):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()

# Pattern 3: Pyramid pattern
print("\nPattern 3:")
for i in range(1, n + 1):
    for k in range(1, n - i + 1):
        print(" ", end="")
    for j in range(1, 2 * i):
      print("*", end="")
    print()


# Pattern 4: Inverted Pyramid pattern
print("\nPattern 4:")
for i in range(n, 0, -1):
    for k in range(1, n - i + 1):
        print(" ", end="")
    for j in range(1, 2 * i):
        print("*", end="")
    print()
```
