```python
#list Comprehensions:
l1=[1,2,3,4]
l2=[3,4,5,6]
sum=list(map(lambda x,y:x+y,l1,l2))
print(sum)
power_list=list(map(lambda x:x**2,sum))
print(power_list)

l3=['sun','butter','sea']
l4=['flower','fly','shore']
concat_list=list(map(lambda x,y:x+'-'.join() +y,l3,l4))
print(concat_list)
```
