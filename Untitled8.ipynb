```python
sentence = "the quick brown fox jumps over the lazy dog even there is a very energetic the dog"
words = sentence.split()
word_count = {word: words.count(word) for word in words}
print(word_count)
l1=[1,2,3,5]
l2=[2,4,5,6]
dic={l1[i]:l2[i] for i in range(len(l1))}

dic
```

```python
l1=[1,2,3,5]
l2=[2,4,5,6]

# Using zip and dict()
dic_zip = dict(zip(l1, l2))
print(dic_zip)
```

```python
l1 = [1, 2, 3, 5]
l2 = [2, 4, 5, 6]

# Using lambda and map to create key-value pairs, then converting to a dictionary
dic_lambda = dict(map(lambda i: (l1[i], l2[i]), range(len(l1))))

print(dic_lambda)
```

```python
import timeit

l1 = list(range(1000))
l2 = list(range(1000, 2000))

# Method 1: Dictionary comprehension
time_comprehension = timeit.timeit(lambda: {l1[i]: l2[i] for i in range(len(l1))}, number=10000)
print(f"Time for dictionary comprehension: {time_comprehension:.6f} seconds")

# Method 2: Using zip and dict()
time_zip_dict = timeit.timeit(lambda: dict(zip(l1, l2)), number=10000)
print(f"Time for zip and dict(): {time_zip_dict:.6f} seconds")

# Method 3: Using lambda and map
time_lambda_map = timeit.timeit(lambda: dict(map(lambda i: (l1[i], l2[i]), range(len(l1)))), number=10000)
print(f"Time for lambda and map: {time_lambda_map:.6f} seconds")
```

```python
import timeit

# Create a large list for comparison
my_list = range(10000)

# Method 1: List comprehension
time_list_comprehension = timeit.timeit(lambda: [x * 2 for x in my_list], number=10000)
print(f"Time for list comprehension: {time_list_comprehension:.6f} seconds")

# Method 2: Using map and lambda
time_map_lambda = timeit.timeit(lambda: list(map(lambda x: x * 2, my_list)), number=10000)
print(f"Time for map and lambda: {time_map_lambda:.6f} seconds")
```
